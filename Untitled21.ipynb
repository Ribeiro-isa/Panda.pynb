```python
import pandas as pd
df = pd.read_csv('/content/sample_data/california_housing_train.csv')

# The column 'Tipo' does not exist in the DataFrame.
# Here are the available columns:
print(df.columns)

# Please replace 'Tipo' with one of the available column names.
# For example, to get value counts for 'housing_median_age':
# df['housing_median_age'].value_counts(normalize=True).to_frame().sort_values('housing_median_age')
```

```python
# Salvando o DataFrame em uma vari√°vel
df_exemplo = df['housing_median_age'].value_counts(normalize=True).to_frame().sort_values('housing_median_age')

df_exemplo.rename(columns={'housing_median_age': 'Percentuais'}, inplace=True)

# Visualizando o DataFrame
df_exemplo
```

```python
display(df.describe())
```

```python
import matplotlib.pyplot as plt

# Ensure the index is a column for plotting if not already
df_plot = df_exemplo.reset_index()

plt.figure(figsize=(15, 7))
plt.bar(df_plot['housing_median_age'], df_plot['proportion'], color='skyblue')
plt.xlabel('Idade Mediana da Casa (housing_median_age)')
plt.ylabel('Percentual (proportion)')
plt.title('Distribui√ß√£o da Idade Mediana da Casa')
plt.xticks(rotation=90) # Rotate x-axis labels if they overlap
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout() # Adjust layout to prevent labels from being cut off
plt.show()
```
